```python
#@title Step 1. Installation: Generate free images from TEXT { display-mode: "form" }
#@markdown 1. Be sure that GPU runtime is selected: Runtime - Change Runtime type --> GPU
#@markdown 2. Click on the play `(>)` button (left from this cell) 
#@markdown
#@markdown When this cell completed successfully (you will get a message below) go to step 2.

#@markdown Credits to: https://huggingface.co/runwayml/stable-diffusion-v1-5

!pip install --upgrade diffusers[torch]
!pip install --upgrade diffusers[flax]
!pip install transformers

try:
  from google.colab import files
  from diffusers import StableDiffusionPipeline
  import torch
  from IPython.display import display
  model_id = "runwayml/stable-diffusion-v1-5"
  pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
  pipe = pipe.to("cuda")
except:
  print("""
  
  
  #########################################
  Runtime will get restarted in 3 seconds
  
  please re-run this cell and be sure that runtime type is GPU.

  #########################################

   """)
  import time
  time.sleep(3)
  import os
  os.kill(os.getpid(), 9)

print("""
#########################################

You are ready to run the next cell, please scroll down to step 2.

#########################################
""")
```

```python
#@title Step 2. Run the program { display-mode: "form" }
#@markdown Describe the image you would like to have
prompt = "a cartoon image an unicorn on the moon" #@param {type:"string"}
#@markdown Filename for your image without spaces, extension or special characters ( _ and - allowed )
image = pipe(prompt).images[0]  
save_as_filename = "unicorn_moon" #@param {type:"string"}
#@markdown The image will be shown below and downloaded to your computer.
#@markdown Click on the play `(>)` button
#@markdown ---
#@markdown Re-run this cell for another image (you do not have to install the application again.)
image.save(f"{save_as_filename}.png")
display(image)
files.download(f"{save_as_filename}.png")
```
